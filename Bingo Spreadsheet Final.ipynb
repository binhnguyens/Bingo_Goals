# Quick program to make a Bingo spreadsheet 

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import matplotlib 
```

# Main

Instructions to use:

- Have a text file saved under `goals.txt`
- Have a list of 24 goals [not numbered and seperated by \n]
- See example of my goals



Changeable parameters
- Entire fig size
- Bingo Size sheet 
- Path to your text file
- Text colour/size

```python
fig, ax = plt.subplots(figsize=(20, 20))

bingo_size = 5

ax.set_xlim(0, bingo_size)
ax.set_ylim(bingo_size, 0)
ax.grid(True)
ax.set_xticks(np.arange(0, bingo_size, 1.0))
ax.set_yticks(np.arange(0, bingo_size, 1.0))
ax.set_aspect('equal', adjustable='box')

# Read lines
with open('goals.txt') as f:
    lines = [line.rstrip() for line in f]
random.shuffle (lines)

# Add the free space
lines.insert(12, 'FREE')

# Create a dictionary for the grid
ds = dict(zip(np.arange (0,bingo_size*bingo_size), lines))

# create tuples of positions
positions = [(x + 0.05, y + 0.5) for x in range(bingo_size) for y in range(bingo_size)]

# Aethetics
colour = '#32466F'

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 12}

matplotlib.rc('font', **font)

for (x, y), (k, v) in zip(positions, ds.items()):
    if (len (v) >= 40):
        ax.text(x, y, f'{v[0:20]}', color=colour)
        ax.text(x, y+0.12, f'{v[20:40]}', color=colour)
        ax.text(x, y+0.24, f'{v[40:len(v)]}', color=colour)
        
    elif (len (v) >= 20):
        ax.text(x, y, f'{v[0:20]}', color=colour)
        ax.text(x, y+0.12, f'{v[20:len(v)]}', color=colour)

    else:
        ax.text(x, y, f'{v}', color=colour)
```
